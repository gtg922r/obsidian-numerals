```python
# Function Parsing

# split mathjs_function_list into a list of lines
mathjs_function_lines = mathjs_function_list.splitlines();
# remove all lines from mathjs_function_lines that contain only the words "Function" and "Description"
mathjs_function_lines = [line for line in mathjs_function_lines if not re.match(r'^\s*Function\s+Description\s*$', line)]

# remove any lines from mathjs_function_lines that don't contain a tab character
mathjs_function_lines = [line for line in mathjs_function_lines if re.search(r'\t', line)]

# make a new list of function names from mathjs_function_lines
mathjs_function_names = [re.sub(r'\(.*$', '', line.split('\t')[0]) for line in mathjs_function_lines]

# if function name contains an = sign, remove everything before and including the = sign (and any whitespace after the = sign)
mathjs_function_names = [re.sub(r'^.*\s*=\s*', '', name) for name in mathjs_function_names]

# remove "math." from the beginning of a function name
mathjs_function_names = [re.sub(r'^math\.', '', name) for name in mathjs_function_names]

mathjs_function_suggestions = [f"f|{name}()" for name in mathjs_function_names];

# print(mathjs_function_suggestions);

```
